```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


# Plot the distribution of satisfaction
sns.countplot(data=airline, x='Satisfaction')
plt.title('Distribution of Satisfaction')
plt.show()


```

# Satisfaction by class

```python
Class_distribution = airline['Class'].value_counts(normalize=True) * 100
Class_distribution
```

```python
plt.figure(figsize=(8, 8))
Class_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Customer Type Distribution')
plt.show()
```

# What are the key drivers of overall customer satisfaction?


```python
correlation_matrix = airline.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# How does customer satisfaction vary by gender?

```python
# Satisfaction by gender
sns.histplot(x='Gender', y='Satisfaction', data=airline)
plt.title('Satisfaction by Gender')
plt.show()


```

# Are there significant differences in satisfaction levels across different age groups?

```python
plt.figure(figsize=(10,6))
sns.histplot(data=airline, x="Age", hue='Satisfaction', kde=True)
plt.show()
```

# How does satisfaction differ between different customer types?



```python
# Satisfaction by customer type
sns.histplot(x='Customer Type', y='Satisfaction', data=airline)
plt.title('Satisfaction by Customer Type')
plt.show()

```

# How do departure and arrival delays impact overall satisfaction?

```python
# Scatter plot of departure delay vs satisfaction
sns.scatterplot(x='Departure Delay', y='Satisfaction', data=airline)
plt.title('Departure Delay vs Satisfaction')
plt.show()


```

```python
# Scatter plot of arrival delay vs satisfaction
sns.scatterplot(x='Arrival Delay', y='Satisfaction', data=airline)
plt.title('Arrival Delay vs Satisfaction')
plt.show()
```
